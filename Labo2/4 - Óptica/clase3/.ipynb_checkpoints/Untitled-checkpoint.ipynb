```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.optimize import curve_fit as cf
from scipy.stats import chi2
from math import floor
```

```python
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.markersize'] = 4
mpl.rcParams['figure.autolayout'] = True
mpl.rcParams['font.size'] = 20
mpl.rcParams['errorbar.capsize'] = 3
mpl.rcParams['lines.markeredgewidth'] = 1
mpl.rcParams['axes.axisbelow'] = True
```

```python
def chi2_red(y,y_err,x,func,popt):
    puntos = len(y)
    params = len(popt)
    grados_lib = puntos - params - 1
    y_ajuste = func(x,*popt)
    
    chi_cuadrado = np.sum(((y-y_ajuste)/y_err)**2)
    chi_reducido = chi_cuadrado/grados_lib
    return chi_reducido

def r2(y,x,lineal,pars):
    r2 = 1 - (np.sum((y-lineal(x,*pars))**2)/np.sum((y - np.mean(y))**2))
    return r2

dcmls = lambda x : -int(floor(np.log10(abs(x))))
lineal = lambda x,a : a*x
```

```python
data1a = np.loadtxt('distancias y altura lente 1.csv',delimiter=',',unpack=True,skiprows=1)
s1a,s_im1a,err_s1a,err_s_im1a,anch1 = data1a
err_anch1 = np.zeros_like(anch1) + 0.1

order1a = np.argsort(s1a)
s1a = s1a[order1a]
s_im1a = s_im1a[order1a]
err_s1a = err_s1a[order1a]
anch1 = anch1[order1a]

plt.errorbar(s1a,s_im1a,yerr=err_s_im1a,fmt='None',ecolor='navy',label='datos',zorder=2)
plt.scatter(s1a,s_im1a,c='navy',zorder=3)
plt.grid()
plt.xlabel(r'$s_{obj}$ [cm]')
plt.ylabel(r'$s_{im}$ [cm]')
plt.legend()
plt.savefig('im_obj1.png')
plt.show()

u1 = 1/s_im1a-1/s1a
err_u1 = np.sqrt((err_s_im1a/s_im1a**2)*2+(err_s1a/s1a**2)**2)
f1 = 1/u1
err_f1 = abs(err_u1/u1**2)

plt.errorbar(s1a,f1,yerr=err_f1,fmt='None',ecolor='navy',label='datos',zorder=2)
plt.scatter(s1a,f1,c='navy',zorder=3)
plt.grid()
plt.xlabel(r'$s_{obj}$ [cm]')
plt.ylabel('f [cm]')
plt.legend()
plt.savefig('foco1.png')
plt.show()

anch01 = 3.1
err_anch01 = 0.1
M1 = anch1/anch01
err_M1 = np.sqrt((err_anch1/anch01)**2+(anch1*err_anch01/anch01**2)**2)

S1 = s_im1a/s1a

order1aa = np.argsort(S1)
S1 = S1[order1aa]
M1 = M1[order1aa]

plt.errorbar(S1[:-1],M1[:-1],yerr=err_M1[:-1],fmt='None',ecolor='navy',label='datos',zorder=2)
plt.scatter(S1[:-1],M1[:-1],c='navy',zorder=3)
plt.grid()
plt.xlabel(r'$s_{im}/s_{obj}$')
plt.ylabel('Aumento')
plt.legend()
plt.savefig('aumento1.png')
```

```python
data1b = np.loadtxt('intensidad lente 1.csv',delimiter=',',unpack=True,skiprows=1)
s1b,s_im1b,i1,err_i1 = data1b
err_sb = np.zeros_like(s1b) + 0.1
ruido = 3.6

order1b = np.argsort(s1b)
s1b = s1b[order1b]
s_im1b = s_im1b[order1b]
i1 = i1[order1b]
err_i1 = err_i1[order1b]

for i in range(len(err_i1)):
    if err_i1[i] < ruido:
        err_i1[i] = ruido

plt.errorbar(s1b,i1,yerr=err_i1,fmt='None',ecolor='Navy',label='datos',zorder=2)
plt.scatter(s1b,i1,c='navy')
plt.grid()
plt.xlabel(r'$s_{obj}$ [cm]')
plt.ylabel('Intensidad [lx]')
plt.legend()
plt.savefig('intensidad1.png')
```

```python
data2 = np.loadtxt('lente 2.csv',delimiter=',',unpack=True,skiprows=1)
s2,s_im2,err_s2,i2,err_i2,anch2,err_anch2 = data2

order2 = np.argsort(s2)
s2 = s2[order2]
s_im2 = s_im2[order2]
err_s2 = err_s2[order2]
i2 = i2[order2]
err_i2 = err_i2[order2]
anch2 = anch2[order2]
err_anch2 = err_anch2[order2]

plt.errorbar(s2,s_im2,yerr=err_s2,fmt='None',ecolor='navy',label='datos',zorder=2)
plt.scatter(s2,s_im2,c='navy',zorder=3)
plt.grid()
plt.xlabel(r'$s_{obj}$ [cm]')
plt.ylabel(r'$s_{im}$ [cm]')
plt.legend()
plt.savefig('im_obj2.png')
plt.show()

u2 = 1/s_im2-1/s2
err_u2 = np.sqrt((err_s2/s_im2**2)*2+(err_s2/s2**2)**2)
f2 = 1/u2
err_f2 = abs(err_u2/u2**2)

plt.errorbar(s2,f2,yerr=err_f2,fmt='None',ecolor='navy',label='datos',zorder=2)
plt.scatter(s2,f2,c='navy',zorder=3)
plt.grid()
plt.xlabel(r'$s_{obj}$ [cm]')
plt.ylabel('f [cm]')
plt.legend()
plt.savefig('foco2.png')
plt.show()

anch02 = 3.1
err_anch02 = 0.1
M2 = anch2/anch02
err_M2 = np.sqrt((err_anch2/anch02)**2+(anch2*err_anch02/anch02**2)**2)

S2 = s_im2/s2

order2aa = np.argsort(S2)
S2 = S2[order2aa]
M2 = M2[order2aa]

plt.errorbar(S2[:-1],M2[:-1],yerr=err_M2[:-1],fmt='None',ecolor='navy',label='datos',zorder=2)
plt.scatter(S2[:-1],M2[:-1],c='navy',zorder=3)
plt.grid()
plt.xlabel(r'$s_{im}/s_{obj}$')
plt.ylabel('Aumento')
plt.legend()
plt.savefig('aumento2.png')
plt.show()


for i in range(len(err_i2)):
    if err_i2[i] < ruido:
        err_i2[i] = ruido

plt.errorbar(s2,i2,yerr=err_i2,fmt='None',ecolor='Navy',label='datos',zorder=2)
plt.scatter(s2,i2,c='navy')
plt.grid()
plt.xlabel(r'$s_{obj}$ [cm]')
plt.ylabel('Intensidad [lx]')
plt.savefig('intensidad2.png')
```

```python
data3 = np.loadtxt('concha diafragma.csv',delimiter=',',unpack=True,skiprows=1)
d,err_d,i3,err_i3 = data3

A = np.pi*(d/2)**2
err_A = np.pi*d*err_d

lineal_b = lambda x,a,b : a*x+b

pars3,cov3 = cf(lineal_b,A,i3,sigma=err_i3,absolute_sigma=True)
a3,b3 = pars3
x_adj3 = np.linspace(min(A),max(A),100)
y_adj3 = lineal_b(x_adj3,a3,b3)

print(chi2_red(i3,err_i3,A,lineal_b,pars3))
print(r2(i3,A,lineal_b,pars3))

plt.plot(x_adj3,y_adj3,'crimson',label='ajuste',zorder=1)
plt.errorbar(A,i3,yerr=err_i3,fmt='None',ecolor='Navy',label='datos',zorder=2)
plt.scatter(A,i3,c='Navy',zorder=3)
plt.grid()
plt.xlabel('rea [cm]')
plt.ylabel('Intensidad [lx]')
plt.legend()
plt.savefig('diafragma.png')
```
