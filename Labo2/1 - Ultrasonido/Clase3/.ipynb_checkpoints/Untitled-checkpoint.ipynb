```python
# funciones para ajustar
#sinc_adj = lambda x,a,b,c : a*np.sinc(b*x)+c
quad = lambda x,a,b,c : a*x**2+b*x+c
```

```python
data1 = np.loadtxt('Emisor1.csv',delimiter=',',unpack=True,skiprows=1) # voltaje según la posición del receptor para el primer emisor
pos1 = data1[0] # posición del receptor
amp1 = data1[3]/2 # amplitud de la onda
phi1 = np.radians(data1[6]) # diferencia de fase entre la onda emitida y medida 
err_pos1 = data1[8] # error de la posición
err_amp1 = data1[9]/2 # error de la amplitud
err_phi1 = np.radians(data1[7]) # error de la fase

#pars1,cov1 = cf(func,x1,y1,p0=[])
#a1,b1,c1 = pars1
#x1_adj = np.linspace(min(pos1),max(pos1),100)
#y1_adj = func(x1_adj,a1,b1,c1)
#res1 = func(pos1,a1,b1,c1)-amp1
#r2_1 = 1 - (np.sum(res1**2)/np.sum((amp1-np.mean(amp1))**2))

print(phi1)

plt.errorbar(pos1,amp1,yerr=err_amp1,xerr=err_pos1,fmt='k.',ecolor='k',label='valores')
#plt.plot(x1_adj,y1_adj,'p',label='ajuste')
plt.xlabel('Posición [m]')
plt.ylabel('Voltaje [mV]')
plt.legend()
plt.grid()
plt.savefig('pos_amp_1.png')
plt.show()
plt.errorbar(pos1,phi1,yerr=err_phi1,xerr=err_pos1,fmt='k.',ecolor='k',label='valores')
plt.xlabel('Posición [mm]')
plt.ylabel('Fase')
plt.legend()
plt.grid()
plt.savefig('pos_phi_1.png')
```

```python
data2 = np.loadtxt('Emisor2.csv',delimiter=',',unpack=True,skiprows=1) # voltaje según la posición del receptor para el primer emisor
pos2 = 340-data2[4] # posición del receptor
amp2 = data2[0]/2 # amplitud de la onda
phi2 = np.radians(data2[2]) # diferencia de fase entre la onda emitida y medida 
err_pos2 = data2[5] # error de la posición
err_amp2 = data2[1]/2 # error de la amplitud
err_phi2 = np.radians(data2[3]) # error de la fase

pars2,cov2 = cf(quad,pos2,amp2)
a2,b2,c2 = pars2
x2_adj = np.linspace(min(pos2),max(pos2),100)
y2_adj = quad(x2_adj,a2,b2,c2)
res2 = quad(pos2,a2,b2,c2)-amp2
r2_2 = 1 - (np.sum(res2**2)/np.sum((amp2-np.mean(amp2))**2))

print(np.round(r2_2,3))

plt.errorbar(pos2,amp2,yerr=err_amp2,xerr=err_pos2,fmt='k.',ecolor='k',label='valores')
plt.plot(x2_adj,y2_adj,'purple',label='ajuste')
plt.xlabel('Posición [mm]')
plt.ylabel('Voltaje [mV]')
plt.legend()
plt.grid()
plt.savefig('pos_amp_2.png')
plt.show()
plt.errorbar(pos2,phi2,yerr=err_phi2,xerr=err_pos2,fmt='k.',ecolor='k',label='valores')
plt.xlabel('Posición [m]')
plt.ylabel('Fase')
plt.legend()
plt.grid()
plt.savefig('pos_phi_2.png')
```
