```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

data = np.loadtxt('ej3.txt')
```

```python
xavg = np.mean(data)
xstd = np.std(data)
print(xavg,xstd)
```

```python
def gauss(x,sig,mu):
    y = np.exp(-0.5*(((x-mu)/sig)**2))/(sig*np.sqrt(2*np.pi))
    return y

sig,mu = curve_fit(gauss,data,np.ones_like(data))[0]

x_adj = np.linspace(min(data),max(data),100)

plt.plot(x_adj,gauss(x_adj,sig,mu),'r')
```

```python
n,bins,patches = plt.hist(data,bins='auto',edgecolor='k',facecolor='purple',alpha=0.75,density=True)

x_hist = []
y_hist = []
for i in range(len(n)):
    x_hist.append((bins[i]+bins[i+1])/2)
    y_hist.append(n[i])
plt.plot(x_hist,y_hist,'go')
```

```python
sig_h,mu_h = curve_fit(gauss,x_hist,y_hist)[0]

x_adj_h = np.linspace(min(data),max(data),100)

plt.plot(x_adj_h,gauss(x_adj_h,sig_h,mu_h),'r')
plt.hist(data,bins='auto',edgecolor='k',facecolor='purple',alpha=0.75,density=True)
```
