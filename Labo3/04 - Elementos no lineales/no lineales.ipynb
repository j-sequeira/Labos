```python
fig,axs = plt.subplots(2,1,figsize=(15,5),layout='constrained')

props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)

df = pd.read_csv(f'Rectificador/100.csv')
fig.suptitle('Voltaje en un rectificador de media onda')
axs[0].plot(df.t,df.V_R,'crimson',zorder=1,label='V\u2091')
axs[0].plot(df.t,df.V_out,'navy',zorder=2,label='V\u2098')
axs[0].set_xlim(min(df.t),max(df.t))
axs[0].set_ylim(min(df.V_R)-1,max(df.V_R)+1)
axs[0].set_xlabel('t [s]')
axs[0].set_ylabel('V [V]')
axs[0].text(0.04,-5,f'R = 100 \u03A9',transform=axs[0].transData,bbox=props,fontsize=13)
axs[0].legend(loc=2,fontsize=13)
axs[0].grid()

df = pd.read_csv(f'Rectificador/30.0.csv')
axs[1].plot(df.t,df.V_R,'crimson',zorder=1,label='V\u2091')
axs[1].plot(df.t,df.V_out,'navy',zorder=2,label='V\u2098')
axs[1].set_xlim(min(df.t),max(df.t))
axs[1].set_ylim(min(df.V_R)-1,max(df.V_R)+1)
axs[1].set_xlabel('t [s]')
axs[1].set_ylabel('V [V]')
axs[1].text(0.04,-5,f'R = 8.2 M\u03A9',transform=axs[1].transData,bbox=props,fontsize=13)
axs[1].legend(loc=2,fontsize=13)
axs[1].grid()

plt.savefig('G4_fig_17oct.pdf',bbox_inches='tight')
```
