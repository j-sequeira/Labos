```python
directorypath="/home/luciano/Escritorio/Labo 3/Diodos/Rectificador"
files = os.listdir(directorypath)
names = []
for i in range(len(files)):
    if files[i].endswith('.csv'):
        names.append(float(files[i][:-4]))
names = np.array(names)
def f(n):
    df4 = pd.read_csv(f'/home/luciano/Escritorio/Labo 3/Diodos/Rectificador/{int(names[n])}.csv')
    vm=df4.V_out
    ve=df4.V_R
    t=df4.t
    plt.plot(t,vm)
    plt.plot(t,ve)
    plt.title(f"R={names[n]}")
    plt.grid()
    plt.savefig(f'/home/luciano/Escritorio/Labo 3/Diodos/Rectificador/{int(names[n])}.png')

    plt.show()
    return
```

```python
for i in range(len(names)):
    f(i)
```
