```python
files = ['100.csv','500.csv','1000.csv','1500.csv','2000.csv','2500.csv','3000.csv','3500.csv','4000.csv','5000.csv','6000.csv','7000.csv','8000.csv','9000.csv','10000.csv',
         '8.2 M.csv']

def func(x,x0,k1,k2,A1,B):
    A2 = A1*np.exp(k2*x0)/(np.exp(-k2/k1*np.pi/2)-np.exp(-k2/k1*2*np.pi))
    C = -A2*np.exp(-k2*(x0+2*np.pi/k1))
    
    res = []
    for y in x:
        if np.mod(k1*(y-x0),2*np.pi) <= np.pi/2:
            res.append(abs(A1)*np.sin(k1*(y-x0))+B)
        else:
            res.append(A2*np.exp(-k2*y)+B+C)
    return np.array(res)

def func(x, x0, k1, k2, A1, B):
    A2 = A1 * np.exp(k2 * x0) / (np.exp(-k2 / k1 * np.pi / 2) - np.exp(-k2 / k1 * 2 * np.pi))
    C = -A2 * np.exp(-k2 * (x0 + 2 * np.pi / k1))

    res = []
    for y in x:
        sine_period = 2 * np.pi / k1
        exp_period = sine_period  # Make the exponential part periodic with the same period as the sine

        y_mod = np.mod(y - x0, sine_period)
        if y_mod <= np.pi / 2:
            res.append(abs(A1) * np.sin(k1 * y_mod) + B)
        else:
            exp_mod = np.mod(y - x0, exp_period)
            A2_mod = A1 * np.exp(k2 * x0) / (np.exp(-k2 / k1 * np.pi / 2) - np.exp(-k2 / k1 * 2 * np.pi))
            C_mod = -A2_mod * np.exp(-k2 * (x0 + 2 * np.pi / k1))
            res.append(A2_mod * np.exp(-k2 * exp_mod) + B + C_mod)

    return np.array(res)
```

```python
for file in files:
    df = pd.read_csv(f'Rectificador/{file}')

    pars,cov = cf(func,df.t,df.V_out,p0=[-0.035,2*np.pi*30,1,max(df.V_out)-min(df.V_out),min(df.V_out)])
    x_adj = np.linspace(min(df.t),max(df.t),100)
    y_adj = func(x_adj,*pars)
    
#    plt.plot(df.t,df.V_R,'crimson',zorder=1,label='V\u2091')
    plt.plot(df.t,df.V_out,'navy',zorder=2,label='V\u2098')
    plt.plot(x_adj,y_adj,'crimson',zorder=1,label='Ajuste')
#       plt.scatter(df.V_R,df.V_out)
    plt.xlim(min(df.t),max(df.t))
#    plt.ylim(min(df.V_R)-1,max(df.V_R)+1)
    plt.xlabel('t [s]')
    plt.ylabel('V [V]')
    plt.title(f'R = {file[:-4]} \u03A9')
    if file == '8.2 M.csv':
        plt.title(f'R = {file[:-4]}\u03A9')
#    plt.legend(loc=2,fontsize=13)
    plt.grid()
    plt.show()
```
