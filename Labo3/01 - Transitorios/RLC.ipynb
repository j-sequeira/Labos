```python
c_list = []

files = os.listdir('RLC sobre')

start_color = mcolors.to_rgb('maroon')
end_color = mcolors.to_rgb('lightsalmon')
num_steps = len(files)

for step in range(num_steps):
    r = np.linspace(start_color[0], end_color[0], num_steps)[step]
    g = np.linspace(start_color[1], end_color[1], num_steps)[step]
    b = np.linspace(start_color[2], end_color[2], num_steps)[step]
    c_list.append((r, g, b))

fig, axs = plt.subplots(len(files),1,figsize=(12,6),sharey=True)

axs[0].set_title('Voltaje medido para distintas resistencias')

for i in range(len(files)):
    if files[i].endswith('.csv'):
        df = pd.read_csv(f'RLC sobre/{files[i]}')
        axs[i].plot(df.t,df.V_m,c=c_list[i],label=f'{files[i][:-4]} \u03A9')
        axs[i].set_xlim(min(df['t']),max(df['t']))
        axs[i].legend(loc=2,fontsize=8)

for a in axs:
    labels = a.get_xticklabels()
    labels[0] = ""
    a.set_xticklabels(labels)

axs[i].set_xlabel('t [s]',fontsize=15)
axs[3].set_ylabel('$V_{m}$ [V]',fontsize=15)
plt.savefig('G4 12-9.pdf')
```
