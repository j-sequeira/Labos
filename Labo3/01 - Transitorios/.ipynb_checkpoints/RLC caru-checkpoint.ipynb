```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.signal import find_peaks
```

```python
directorypath="C:/Users/LCB.DESKTOP-CJ4NPO0/Downloads/RLC sobre"
files = os.listdir(directorypath)
names = []
for i in range(len(files)):
    if files[i].endswith('.csv'):
        names.append(float(files[i][:-4]))
names = np.array(names)
def f(n):
    df = pd.read_csv(f'C:/Users/LCB.DESKTOP-CJ4NPO0/Downloads/RLC sobre/{names[n]}.csv')
    vm=df.V_m
    ve=df.V_e
    t=df.t
    # plt.plot(t,vm)
    # plt.plot(t,ve)
    return t,vm,ve
```

```python
n=4
t=f(n)[0]
vm=f(n)[1]
ve=f(n)[2]
plt.plot(t,vm)
plt.plot(t,ve)
plt.show()
max(vm)
k=1500
b=np.argmin(ve)
h=np.argmax(vm)
tpp= t[h:]
tp=tpp[:k]
vmpp= vm[h:]
vmp=vmpp[:k]

def seno_subamortiguado(t, A, omega, phi, tau,b):
    return 5 * np.exp(-t / tau) * np.sin(omega * (t+phi)+b )+4

# Ajusta la funcin a los datos
parametros_iniciales = [3,100000,0,0.009,0]  # Adivinanza inicial
parametros_optimizados, _ = curve_fit(seno_subamortiguado, tp, vmp, p0=parametros_iniciales,maxfev=100000)

# Grafica los datos y la funcin ajustada
plt.figure(figsize=(8, 6))
plt.plot(tp, vmp, label='Datos experimentales', color='blue', marker='.')
plt.plot(tp, seno_subamortiguado(tp, *parametros_optimizados), label='Ajuste', color='red')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [v]')
plt.legend()
plt.grid(True)
# plt.savefig(C:\Users\LCB.DESKTOP-CJ4NPO0\Downloads\RLC sobre',)
plt.show()
print('A=',parametros_optimizados[1])
print('tau=',parametros_optimizados[3])
```
