```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import os
import pandas as pd
from math import floor
from scipy.optimize import curve_fit as cf
import scipy.stats as stats
```

```python
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.markersize'] = 4
mpl.rcParams['figure.autolayout'] = True
mpl.rcParams['font.size'] = 15
mpl.rcParams['errorbar.capsize'] = 3
mpl.rcParams['lines.markeredgewidth'] = 1
mpl.rcParams['axes.axisbelow'] = True
```

```python
mus = [1,2,4,8]

fig,axs = plt.subplots(2,2,figsize=(16,9),layout='tight')
axs = axs.reshape(-1)

for i in range(len(mus)):
    data = np.random.poisson(mus[i],5000)
    axs[i].hist(data,bins=np.arange(-0.5,21.5),density=True,color='navy',edgecolor='black',label=f'<m> = {mus[i]}')
    axs[i].set_xlabel('m')
    axs[i].set_ylabel('Pm(T)')
    axs[i].grid()
    axs[i].legend()

plt.savefig('plots/poisson_ej.png')
plt.show()
plt.close()
```

```python
be_dist = lambda x,m : m**x/(1+m)**(x+1)
c_list = ['navy','forestgreen','darkmagenta','sienna','tab:orange']
```

```python
values = np.arange(1,50)
mus = [1,2,4,8]

fig,axs = plt.subplots(2,2,figsize=(16,9),layout='tight')
axs = axs.reshape(-1)

for i in range(len(mus)):
    data = []
    unnormalized_probabilities = be_dist(values,1)
    probabilities = unnormalized_probabilities / np.sum(unnormalized_probabilities)
    for j in range(5000):
        data.append(np.random.choice(values, p=probabilities))
    data = np.array(data)
    
    axs[i].hist(data,bins=np.arange(-0.5,21.5),density=True,color='navy',edgecolor='black',label=f'<m> = {mus[i]}')
    axs[i].set_xlabel('m')
    axs[i].set_ylabel('Pm(T)')
    axs[i].grid()
    axs[i].legend()

plt.savefig('plots/BE_ej.png')
plt.show()
plt.close()
```
